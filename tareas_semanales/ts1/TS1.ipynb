Considerando que por la pata negativa existe una tension que la llamare $V_x$ que por corto virtual tambien estara por la pata positiva, de esta manera se puede deducir que $V_x$ se comporta como un divisor de tension por lo que se puede obtener la siguiente igualdad.

$$V_x = V_1  \frac{R_3}{\frac{1}{SC_1}+R_3} -----> V_x = V_1  \frac{SC_1R_3}{1+SC_1R_3}$$

$$\frac{V_1-V_x}{R_1} = \frac{V_x-V_2}{R_2} -----> \frac{V_1}{R_1} + \frac{V_2}{R_2} = V_x(\frac{1}{R_2} + \frac{1}{R_1}) ----> \frac{V_1}{R_1} + \frac{V_2}{R_2} = V_x(\frac{R_1 + R_2}{R_1R_2}) $$

Reemplazando valores:

$$ (\frac{V_1}{R_1} + \frac{V_2}{R_2})(\frac{R_1R_2}{R_1 + R_2}) = V_1\frac{SC_1R_3}{SC_1R_3 + 1}$$

$$ V_1\frac{R_2}{R_1 + R_2} - V_1\frac{SC_1R_3}{SC_1R_3 + 1} = -V_2\frac{R_1}{R_1 + R_2}$$

y operando un poco y reemplazando valores se llega a la siguiente transformada

$$ T(s) = \frac{V_2}{V_1} = \frac{SC_1R_3R_1 - R_2}{SC_1R_3R_1 + R_1} ----> T(s) = \frac{S-\frac{R_2}{C_1R_3R_1}}{S+\frac{1}{C_1R_3}}$$

Si se toma a $\frac{R_2}{R_1} = 1$ se llega a la siguiente ecuacion

$$T(s) = \frac{S-\frac{1}{C_1R_3}}{S+\frac{1}{C_1R_3}}$$

## punto 3

IMPEDANCIA NORMALIZADA

$$T(s) = \frac{SC_1R_3 - 1}{SC_1R_3 + 1} ---> T(s) = \frac{\frac{Z_{R3}}{Z_{C1}} - 1}{\frac{Z_{R3}}{Z_{C1}} + 1}$$

normalizando a $R_3 = 1 \ohm $ y $C_1 = 1 F$ nos queda de la siguiente manera

$$T(s) = \frac{S - 1}{S + 1}$$

FRECUENCIA NORAMLIZADA

Para normalizarla en frecuencia analizo cual es la frecuencia de corte de la transferencia.

$$T(s) = \frac{S-\frac{1}{C_1R_3}}{S+\frac{1}{C_1R_3}} ---> Wo = \frac{1}{C_1R_3}$$

Entonces normalizando a $S' = S/Wo$

$$ T(s) = \frac{SC_1R_3-1}{SC_1R_3+1} ---> T(s) = \frac{\frac{S}{W_o}-1}{\frac{S}{W_o} + 1} $$

$$T(s') = \frac{S'-1}{S'+1}$$



## punto 4

Simulacion de la transferencia normalizada

$$T(s) = \frac{S-1}{S+1}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sp

from scipy.signal import TransferFunction as tf
from pytc2.sistemas_lineales import pzmap

num = [1, -1]
den = [1, 1]

#DIAGRAMA DE POLOS Y CEROS
T = tf(num, den)
pzmap(T)


#MODULO Y FASE
w, h = sp.freqs(num, den)

plt.figure(figsize=(12,6))
# Magnitud en dB
plt.subplot(1,2,1)
plt.semilogx(w, 20 * np.log10(np.abs(h)))
plt.ylim([-10, 1])
plt.ylabel("Magnitud [dB]")
plt.xlabel("Frecuencia [rad/s]")
plt.grid(which="both")

# Fase en grados
plt.subplot(1,2,2)
plt.semilogx(w, np.angle(h, deg=True))
plt.ylabel("Fase [°]")
plt.xlabel("Frecuencia [rad/s]")
plt.grid(which="both")

plt.tight_layout()
plt.show()
```
